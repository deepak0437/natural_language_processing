import the library

```python
import pandas as pd
import numpy as np
```

import the dataset

```python
df = pd.read_csv('Stock_prediction.csv', encoding = "ISO-8859-1")
df
```

divide dataset into train and test set

```python
train = df[df['Date'] < '20150101'] ##training set are less that 1st jan 2015
test = df[df['Date'] > '20141231']  ###testing set are greater than 31st dec 2014
```

Removing all the punctuations

```python
data = train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ", regex=True, inplace=True)
data.head(1)
```

Removing the column name all put o to 25 for easy access

```python
list1 = [i for i in range(25)]   
new_index = [str(i) for i in list1]
data.columns = new_index
data.head(1)
```

converts all the texts into lower case

```python
for index in new_index:
  data[index] = data[index].str.lower()
data.head(1)
```

join very first paragraph of all 25 review which is given

```python
' '.join(str(x) for x in data.iloc[0,0:25])
```

join all paragraph of the dataset

```python
headline = []
for row in range(len(data.index)):
  headline.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headline[4]
```

import library of bag of words and random classifier

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector = CountVectorizer(ngram_range=(2,2))
traindataset = countvector.fit_transform(headline) ##fit transform the dataset
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
