import one_hot library

```python
from tensorflow.keras.preprocessing.text import one_hot
```

Datasets or sentences

```python
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

vocabulary size in 10000

```python
##vocab_size means actually what is the size of dictonary
voc_size = 10000
```

one hot representation

```python
##coverting sentences into one hot representation 
onehot_repr = [one_hot(words, voc_size) for words in sent]
print(onehot_repr)
##print one hot representation to provide the index of each word in sentences from dictonary(vocab_size).
```

import library for embedding

```python
import numpy as np
from tensorflow.keras.layers import Embedding   ##whatever one hot representaion we have passing through embedding layers in keras to form a embedding matrix
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential  ##to create a embedding layer
```

```python
sent_length = 8  ##length of sentecese what we want 
embedded_docs = pad_sequences(onehot_repr, padding='pre', maxlen= sent_length)
print(embedded_docs)
```

```python
dim = 10  ##giving value of feature value or dimension 
```

create and compile the embedding model

```python
model = Sequential()
model.add(Embedding(voc_size, 10, input_length=sent_length))
model.compile('adam', 'mse')   ##mean squared error
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
