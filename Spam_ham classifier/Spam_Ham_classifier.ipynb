import library

```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import re
```

import the dataset

```python
dataset = pd.read_csv('SMSSpamCollection', sep= '\t', names = ['label', 'message'])
dataset                         ##names is declared the column name of dataset i.e label and message
```

decleared the variable of module stemming and lemmatizing

```python
stemmer = PorterStemmer()
wordnet = WordNetLemmatizer()
```

cleaning the texts or datsets.

```python
corpus = []
for x in range(len(dataset)):
  review = re.sub('[^a-zA-Z]', ' ', dataset['message'][x])
  review = review.lower()
  review = review.split()
  review = [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
  review = ' '.join(review)
  corpus.append(review)
  print([corpus[x]], len(corpus))
```

create bag of words model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#from sklearn.feature_extraction.text import CountVectorizer
cv = TfidfVectorizer()
x = cv.fit_transform(corpus).toarray()
```

```python
print(x[1])
print(x.shape)
```

decleared dependet(y) value and independent(x) value

```python
y = pd.get_dummies(dataset['label'])
y
```

```python
y = y.iloc[:,1].values
y
```

```python
x = cv.fit_transform(corpus).toarray()
x
```

train, test and split the model

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

training the model using naive bayes classifier

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train, y_train)
```

predicting the model

```python
y_pred = model.predict(x_test)
y_pred
```

check the accuracy of predicton

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
```
